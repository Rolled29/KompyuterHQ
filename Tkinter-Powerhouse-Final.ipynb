# This Function Converts a csv file into a multiindex Dataframe

```python
    def store_csv(store):
        df = pd.read_csv(store, skiprows=1)
        cols = pd.MultiIndex.from_tuples([("MOTHERBOARD","BRAND"), ("MOTHERBOARD","NAME"), ("MOTHERBOARD",'PRICE'),("MOTHERBOARD",'STOCKS'), ("MOTHERBOARD",'USER RATING'),
                                         ("CPU","BRAND"), ("CPU","NAME"), ("CPU",'PRICE'),("CPU",'STOCKS'), ("CPU",'USER RATING'),
                                         ("CPU COOLER","BRAND"), ("CPU COOLER","NAME"), ("CPU COOLER",'PRICE'),("CPU COOLER",'STOCKS'), ("CPU COOLER","USER RATING"),
                                         ("GPU","BRAND"),("GPU","NAME"), ("GPU",'PRICE'),("GPU",'STOCKS'), ("GPU","USER RATING"),
                                         ("RAM","BRAND"),("RAM","NAME"), ("RAM",'PRICE'),("RAM",'STOCKS'), ("RAM","USER RATING"),
                                         ("PSU","BRAND"), ("PSU","NAME"), ("PSU",'PRICE'),("PSU",'STOCKS'), ("PSU","USER RATING"),
                                         ("STORAGE","BRAND"),("STORAGE","NAME"), ("STORAGE",'PRICE'),("STORAGE",'STOCKS'),("STORAGE","USER RATING"),
                                         ("CHASIS","BRAND"),("CHASIS","NAME"), ("CHASIS",'PRICE'),("CHASIS",'STOCKS'),("CHASIS","USER RATING")
                                         ])
        df.columns = cols
        return df
```

# This function creates another window and displays the graph

```python
def display_graph(store,part, title,row, column):
    graph_window =Toplevel(root)
    graph_window.iconbitmap('icon.ico')
    df_graph = store_csv(store + '.csv')
    brands = []
    x = df_graph[part]['BRAND'].tolist()
    for brand in x:
        if brand not in brands:
            brands.append(brand)
    avg_list = []
    for i in brands:
        avg_list.append(df_graph[df_graph[part]['BRAND'] == i][part]['USER RATING'].mean())
    figure = plt.figure(figsize=(8, 10))
    plt.title(title)
    plt.bar(x=brands, height=avg_list)
    plt.xticks(x, rotation=60)
    bar = FigureCanvasTkAgg(figure, graph_window)
    bar.get_tk_widget().grid(row = row, column = column)
```

# This function gets the price of the selected part and change the price currently displayed

```python
def getprice(store, component, name,price_display,rating_display):
    global mobo_price,mobo_rating,cpu_price,cpu_rating,cpu_cooler_price,cpu_cooler_rating,gpu_price, gpu_rating,ram_price,ram_rating,psu_price, psu_rating,storage_price,storage_rating,case_price, case_rating
    price = 0
    rating = 0
    df = store_csv(store + '.csv')
    for i in range (len(df)):
        if df[component]['NAME'][i] == name:
            price = df[component]['PRICE'][i]
            rating = df[component]['USER RATING'][i]
    if (component == 'MOTHERBOARD'):
        mobo_price = price
        mobo_rating = rating
    elif (component == 'CPU'):
          cpu_price = price
          cpu_rating = rating
    elif (component == 'CPU COOLER'):
         cpu_cooler_price = price
         cpu_cooler_rating = rating
    elif (component == 'GPU'):
         gpu_price = price
         gpu_rating = rating
    elif (component == 'RAM'):
         ram_price = price
         ram_rating = rating
    elif (component == 'PSU'):
         psu_price = price
         psu_rating = rating
    elif (component == 'STORAGE'):
          storage_price = price
          storage_rating = rating
    elif (component == 'CHASIS'):
          case_price = price
          case_rating = rating
    price_display.config(text= str(price) + 'php')
    rating_display.config(text=rating)
    total_cost = mobo_price + cpu_price + cpu_cooler_price + gpu_price + ram_price + psu_price + storage_price + case_price
    system_builder_frame.display_total.grid_forget()
    system_builder_frame.display_total = Label(frame_body, text =  str(total_cost)+'php', font = 'Arial 12 bold', bg = 'white')
    system_builder_frame.display_total.grid(row = 9, column =4)

    
```

# This function changes the choices in dropdown list based on the store selected

```python
def store_eventlistener (store,part, part_select, menu, row, column):
    if (store == 'EasyPC'):df = store_csv('EasyPC.csv')
    elif(store == 'Kedem Computers'):df = store_csv('Kedem Computers.csv')
    elif(store == 'Dynaquest'):df = store_csv('Dynaquest.csv') 
    menu = OptionMenu(frame_body, part_select, *df[part]['NAME'].values.tolist())
    menu.config(width=57)
    menu.configure(activebackground="#575656")
    menu.grid(row = row, column =column)
```

# This function pop ups a new window for the price list selected

```python
def popup_pricelist(part, store):
    stores  = ['EasyPC', 'Kedem Computers', 'Dynaquest']
    newWindow =Toplevel(root)
    newWindow.iconbitmap('icon.ico')
    df = store_csv(stores[store] + '.csv')
    x = df[part]
    newWindow.geometry("900x420") 
    newWindow.title(stores[store] + ' Price List-' + part)
    frame = Frame(newWindow)
    frame.pack(fill='both', expand=True)
    pt = Table(frame,dataframe = x, showtoolbar=True, showstatusbar=True, cellwidth = 130)
    pt.show()    
```
