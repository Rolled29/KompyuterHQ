```python
import sqlite3
try:
    conn = sqlite3.connect('Powerhouse.db')
    sqlite_create_table_motherboard = '''CREATE TABLE MOTHERBOARD (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_cpu = '''CREATE TABLE CPU (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_cpucooler = '''CREATE TABLE CPU_COOLER (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_gpu = '''CREATE TABLE GPU (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_ram = '''CREATE TABLE RAM (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_psu = '''CREATE TABLE PSU (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_storage = '''CREATE TABLE STORAGE (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
    sqlite_create_table_chasis = '''CREATE TABLE CHASIS (
                                BRAND TEXT NOT NULL,
                                NAME TEXT PRIMARY KEY,
                                PRICE FLOAT,
                                STOCKS INTEGER,
                                USER RATING FLOAT
                                );
                                '''
 

    cursor = conn.cursor()
    print("Successfully Connected to SQLite")
    cursor.execute(sqlite_create_table_motherboard)
    cursor.execute(sqlite_create_table_cpu)
    cursor.execute(sqlite_create_table_cpucooler)
    cursor.execute(sqlite_create_table_gpu)
    cursor.execute(sqlite_create_table_ram)
    cursor.execute(sqlite_create_table_psu)
    cursor.execute(sqlite_create_table_storage)
    cursor.execute(sqlite_create_table_chasis)
    conn.commit()
    print("SQLite table created")

    cursor.close()

except sqlite3.Error as error:
    print("Error while creating a sqlite table", error)
finally:
    if conn:
        conn.close()
        print("sqlite connection is closed")

```
