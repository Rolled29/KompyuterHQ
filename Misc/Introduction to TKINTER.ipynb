# Controlling Tkinter Application Layout
In order to organize or arrange or place all the widgets in the parent window, Tkinter provides us the geometric configuration of the widgets. The GUI Application Layout is mainly controlled by Geometric Managers of Tkinter.

It is important to note here that each window and Frame in your application is allowed to use only one geometry manager. Also, different frames can use different geometry managers, even if they're already assigned to a frame or window using another geometry manager.

There are mainly three methods in Geometry Managers: pack, grid,place

# Tkinter pack() with Parameters: fill, side and, expand.

The <b>fill</b> argument specify in which direction  the frames should fill. 
<li><b>tk.X</b> fill in the horizontal direction </li>
<li><b>tk.Y</b> is used to fill vertically </li>
<li><b>tk.BOTH</b> fill bith directions</li>

```python
#Using fill parameter for pack()
import tkinter as tk
win= tk.Tk()

frame1 = tk.Frame(master=win, height=80, bg="red")
# adding the fill argument with 
# horizontal fill value
frame1.pack(fill=tk.X)

frame2 = tk.Frame(master=win, height=50, bg="yellow")
frame2.pack(fill=tk.X)

frame3 = tk.Frame(master=win, height=40, bg="blue")
frame3.pack(fill=tk.X)

win.mainloop()
```

```python
#Using all options: fill, side, expand
import tkinter as tk

win = tk.Tk()

frame1 = tk.Frame(master=win, width=200, height=100, bg="Yellow")
# setting fill, side and expand
frame1.pack(fill=tk.BOTH, side=tk.LEFT, expand=True)

frame2 = tk.Frame(master=win, width=100, bg="blue")
frame2.pack(fill=tk.BOTH, side=tk.LEFT, expand=True)

frame3 = tk.Frame(master=win, width=50, bg="green")
frame3.pack(fill=tk.BOTH, side=tk.LEFT, expand=True)

win.mainloop()
```

# Tkinter grid() Geometry Manager

The grid() geometry manager is mainly used to split either a window or frame into rows and columns.

To specify the location of a widget, call the grid() function and pass the row and column indices to the row and column keyword arguments, respectively.

Index of both the row and column starts from 0, so a row index of 2 and a column index of 2 tells the grid() function to place a widget in the third column of the third row(0 is first, 1 is second and 2 means third).

<i>syntax of the grid() function:<i><b><widget.grid(options) </b>

 
<i>The possible options as a parameter to this method are given below</i>

<li><b>Column</b>

This option specifies the column number in which the widget is to be placed. The index of leftmost column is 0.</li>

<li><b>Row</b>

This option specifies the row number in which the widget is to be placed. The topmost row is represented by 0.

<li><b>Columnspan</b>

This option specifies the width of the widget. It mainly represents the number of columns up to which, the column is expanded.

<li><b>Rowspan</b>

This option specifies the height of the widget. It mainly represents the number of rows up to which, the row is expanded.

<li><b>padx, pady</b>

This option mainly represents the number of pixels of padding to be added to the widget just outside the widget's border.

<li><b>ipadx, ipady</b>

This option is mainly used to represents the number of pixels of padding to be added to the widget inside the widget's border.

<li><b>Sticky</b>

If any cell is larger than a widget, then sticky is mainly used to specify the position of the widget inside the cell. It is basically concatenation of the sticky letters which represents the position of the widget. It may be <b> N, E, W, S, NE, NW, NS, EW, ES </b>

```python
#Sample creating 5 x 3 grid
import tkinter as tk

win = tk.Tk()

for i in range(5):
    for j in range(3):
        frame = tk.Frame(
            master = win,
            relief = tk.RAISED,
            borderwidth = 1
        )
        frame.grid(row=i, column=j)
        label = tk.Label(master=frame, text=f"Row {i}\nColumn {j}")
        label.pack()

win.mainloop()
```

# Tkinter place() Geometry Manager
The place() Geometry Manager organizes the widgets to place them in a specific position as directed by the programmer. It organizes the widget in accordance with its x and y coordinates. Both x and y coordinates are in pixels.

Thus the origin (where x and y are both 0) is the top-left corner of the Frame or the window. The y argument specifies the number of pixels of space from the top of the window, to place the widget, and the x argument specifies the number of pixels from the left of the window.

<i>syntax of the place() method:</i> <b>widget.place(options)</b>

<i>The possible options as a parameter to this method are:</i>

<li><b>x, y</b></li>

This option indicates the horizontal and vertical offset in the pixels.

<li><b>bheight, width</b></li>

This option indicates the height and weight of the widget in the pixels.

<li><b>bAnchor</b></li>

This option mainly represents the exact position of the widget within the container. The default value (direction) is NW that is (the upper left corner).

<li><b>bbordermode</b></li>

This option indicates the default value of the border type which is INSIDE and it also refers to ignore the parent's inside the border. The other option is OUTSIDE.

<li><b>brelx, rely</b></li>

This option is used to represent the float between 0.0 and 1.0 and it is the offset in the horizontal and vertical direction.

<li><b>relheight, relwidth</b></li>

This option is used to represent the float value between 0.0 and 1.0 indicating the fraction of the parent's height and width.

# Python Tkinter Label Widget

<li>Tkinter Label widget in Python is used to create a Label in the GUI application in which we can show any text or image.</li>

<li>The label widget in Tkinter is used to display boxes where you can place your images and text.</li>

<li>The label widget is mainly used to provide a message about the other widgets used in the Python Application to the user.</li>

<b>Tkinter Label Widget</b>
The syntax of the label widget is given below,

<b>W = Label(master,options) </b>  
In the above syntax, the master parameter denotes the parent window. You can use many options to configure the text and these options are written as comma-separated key-value pairs.

<b>Tkinter Label Widget Options</b>

<li><b>anchor</b>	This option is mainly used for controlling the position of text in the provided widget size. The default value is CENTER which is used to align the text in center in the provided space.</li>
<li><b>bd</b>	This option is used for the border width of the widget. Its default value is 2 pixels.</li>
<li><b>bitmap</b>	This option is used to set the bitmap equals to the graphical object specified so that now the label can represent the graphics instead of text.</li>
<li><b>bg</b>	This option is used for the background color of the widget.</li>
<li><b>cursor</b>	This option is used to specify what type of cursor to show when the mouse is moved over the label. The default of this option is to use the standard cursor.</li>
<li><b>fg</b>	This option is used to specify the foreground color of the text that is written inside the widget.</li>
<li><b>font</b>	This option specifies the font type of text inside the label.</li>
<li><b>height</b>	This option indicates the height of the widget</li>
<li><b>image</b>	This option indicates the image that is shown as the label.</li>
<li><b>justify</b>	This option specifies the alignment of multiple lines in the label. The default value is CENTER. Other values are RIGHT, LEFT; you can justify according to your requirement</li>
<li><b>padx</b>	This option indicates the horizontal padding of the text. The default value of this option is 1.</li>
<li><b>pady</b>	This option indicates the vertical padding of the text. The default value of this option is 1.</li>
<li><b>relief</b>	This option indicates the type of border. The default value of this option is FLAT</li>
<li><b>text</b>	This option is set to the string variable and it may contain one or more than one line of text</li>
<li><b>textvariable</b>	This option is associated with a Tkinter variable that is (StringVar) with a label. If you change the value of this variable then text inside the label gets updated.</li>
<li><b>underline</b>	This option is used to underline a specific part of the text. The default value of this option =-1(no underline); you can set it to any integer value up to n and counting starts from 0.</li>
<li><b>width</b>	This option indicates the width of the widget.</li>
<li><b>wraplength</b>	Rather than having only one line as the label text, you can just break it to any number of lines where each line has the number of characters specified to this option.</li>

```python
#Sample Label Widget and string variable
import tkinter
from tkinter import *

win = Tk()

var = StringVar() #variable StringVar() and then assigned a value to it, and this variable is assigned as value to the textvariable option of the Label widget.
label = Label( win, textvariable=var, relief=RAISED )

# set label value
var.set("Hey!? Welcome to Tkinter")

label.pack()
win.mainloop()
```

# Python Tkinter Radiobutton Widget

The syntax of the Radiobutton widget is given below:

<b>W = Radiobutton(master, options)   </b>

<h1>Tkinter Radiobutton Widget Methods:</h1>

<li><b>deselect()</b>	This method is used to deselect or turns off the radio button</li>
<li><b>select()</b>	This method is used to select the radio button</li>
<li><b>invoke()</b>	This method is generally used to call a function when the state of radio button gets changed.</li>
<li><b>flash()</b>	This method is generally used to flash the radio button between its normal and active colors many times.</li>

```python
from tkinter import * 
from tkinter.ttk import *

# Creating parent Tkinter window 
win = Tk() 
win.geometry("200x200") 

# let us create a Tkinter string variable 
# that is able to store any string value 
v = StringVar(win, "1") 

# here is a Dictionary to create multiple buttons 
options = {" Option A" : "1", 
"Option B" : "2", 
"Option C" : "3", 
"Option D" : "4" 
} 

# We will use a Loop just to create multiple 
# Radiobuttons instaed of creating each button separately
for (txt, val) in options.items(): 
    Radiobutton(win, text=txt, variable=v, value=val).pack(side = TOP, ipady = 4) 

mainloop() 
```

# Python Tkinter MessageBox

<li>Create and use Tkinter MessageBox while developing desktop applications using the MessageBox module in Tkinter.</li>

<li>The functions of the MessageBox module are as follows: showError(), askretrycancel(), showwarning()</li>

<b>Tkinter MessageBox</b>
To use the messagebox module, we first need to import it in our python script:

<i>from tkinter import messagebox</i>

Then following is the basic syntax to use the messagebox:

<b>messagebox.function_name(title, message [, options]) </b> 


<li><b>function_name</b>

This is used to indicate the name of the appropriate MessageBox Function.</li>

<li><b>title</b>

This is used to indicate the text to be displayed in the title bar of the appropriate message box.</li>

<li><b>message</b>

This is used to indicate the text to be displayed as a message in the message box.</li>

<li><b>options</b>

It is used to indicate various options in order to configure the MessageBox. There are two values of it and these are default and parent.</li>

<li><b>default:</b> It is used to specify a default button such as ABORT, RETRY, IGNORE.</li>

<li><b>parent:</b> It is used to specify a window on the top of which we will display the MessageBox.</li>

The functions present in the MessageBox module uses the same syntax but the functionalities of each function are different.</li>



<b>Tkinter MessageBox - showwarning()</b>
<p>This method is used to display any warning to the user in a Python</p>

```python
#Tkinter MessageBox - showwarning()

#This method is used to display any warning to the user in a Python
from tkinter import *  
  
from tkinter import messagebox  
  
win = Tk()  
win.geometry("200x200")  
messagebox.showwarning("warning","This is a Warning")  
  
win.mainloop()  
  
```

```python
#Tkinter MessageBox - askquestion()
#This method in MessageBox is mainly used to ask some question to the user which can be answered either in Yes or No.
from tkinter import *  
from tkinter import messagebox  
  
win = Tk()  
win.geometry("100x100")  
messagebox.askquestion("Are you sure you want to exit game?")  
win.mainloop()
```

```python
#Tkinter MessageBox - askretrycancel()
#If you want to ask a user to do any particular task or not
from tkinter import *  
from tkinter import messagebox  
  
win= Tk()  
win.geometry("100x100")  
messagebox.askretrycancel("Application"," Try again?")  
  
win.mainloop()  
```
